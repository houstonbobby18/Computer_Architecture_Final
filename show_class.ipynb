```python
import os

from IPython.display import Image
import colorama as cl

import inferencing as inf
#Imports the scripts we will call
```

```python
# Shows the image of the banana pulled from the internet
banana_path = r"images_to_be_proceessed/banana__Web_Picture_1.jpg"
Image(filename=banana_path, width=200, height=200)
```

```python
# Turns the image into a Numpy array, often called preprocessing
numpy_banana = inf.preprocess_image(banana_path)
print(numpy_banana.shape)
```

```python
inferenced_banana = inf.class_example_inference(numpy_banana)
print(inferenced_banana.shape)
```

```python
guessed_fruit = inf.inferencing(numpy_banana)
print(guessed_fruit)
```

```python
dirctory = r"images_to_be_proceessed" # Relative directory to the images

for file in os.listdir(dirctory):
    
    if file.endswith(".jpg"):
        file_path = dirctory + "/" + file
        display(Image(filename=file_path, width=200, height=200))
        fruit_guessed=inf.run_model(file_path)
        actual_fruit = file.split("_")[0]
        if fruit_guessed == actual_fruit:
            # Make print to text green
            print( "Correctly guessed:"+cl.Fore.GREEN +f" {actual_fruit}")
            print(cl.Style.RESET_ALL, end="\n")
        else:
            # Make print to text red
            print("Incorrectly guessed:"+cl.Fore.RED +f" {fruit_guessed}")
            print(cl.Style.RESET_ALL, end="")
            print(f"Actual fruit: {actual_fruit}\n")
```
